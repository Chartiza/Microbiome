## Script is designed to vizualize Abundance-Prevalence plot from your custommade table

```python
# import packages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tax = pd.read_excel('input/My_tax_manual_editing.xlsx')
main = tax.sort_values(by=['Abundance'], ascending=False).reset_index(drop=True)
main.head()
```

```python
plt.figure(figsize=(14, 7))

sns.scatterplot(data=tax, x="Abundance", y="Prevalence", hue="domain")
sns.despine()
plt.legend(loc='center left', bbox_to_anchor=(0.9, 0.40), ncol=1)

for i in range(0,3):
    plt.text(main['Abundance'][i], main['Prevalence'][i]+50, main['sp_abud'][i])
for i in range(3,4):
    plt.text(main['Abundance'][i], main['Prevalence'][i]+120, main['sp_abud'][i])        
for i in range(4,5):
    plt.text(main['Abundance'][i], main['Prevalence'][i]-80, main['sp_abud'][i])
for i in range(5,6):
    plt.text(main['Abundance'][i], main['Prevalence'][i]+30, main['sp_abud'][i])
    
plt.tight_layout()
plt.savefig('results/All_domain.jpg', dpi=150)
```

DEFINE CORE SP

```python
core = tax.loc[tax['Prevalence'] > 1650].sort_values(by=['Abundance'], ascending=False).reset_index(drop=True)
medium = tax.loc[(tax['Prevalence'] <= 1650)&(tax['Prevalence'] >= 200)].reset_index(drop=True)
rare = tax.loc[tax['Prevalence'] < 200].sort_values(by=['Prevalence'], ascending=False).reset_index(drop=True)
```

```python
plt.figure(figsize=(9, 7))

sns.scatterplot(data=core, x="Abundance", y="Prevalence", hue="domain")
sns.despine()
plt.legend(loc='center left', bbox_to_anchor=(0.7, 0.40), ncol=1)
plt.suptitle("Gut metagenome species, precent in 90% smpls"+
            '\n'+str(len(core))+' species')   

for i in range(0,2):
    plt.text(core['Abundance'][i]+0.15, core['Prevalence'][i]+3, core['sp_abud'][i])
for i in range(2,3):
    plt.text(core['Abundance'][i], core['Prevalence'][i]+5, core['sp_abud'][i])
for i in range(3,4):
    plt.text(core['Abundance'][i], core['Prevalence'][i]+15, core['sp_abud'][i])       
for i in range(4,5):
    plt.text(core['Abundance'][i], core['Prevalence'][i]-10, core['sp_abud'][i])
for i in range(5,6):
    plt.text(core['Abundance'][i], core['Prevalence'][i]+5, core['sp_abud'][i])

plt.tight_layout()
plt.savefig('results/Core_domain.jpg', dpi=150)
core.to_excel('results/Core.xls')
```
